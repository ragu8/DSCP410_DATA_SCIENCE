##### AIM:
To write Python program to perform simple arithmetic operations on 2D arrays using
NumPy package.

```python
import numpy as np
```

```python
#this function returns the sum of two matrices 
def fn_mat_sum(mat_a, mat_b):
    if mat_a.shape == mat_b.shape:
        mat_sum = mat_a + mat_b
    else:
        mat_sum = None
    return mat_sum
```

```python
#This function returns the difference of two matrices
def fn_mat_diff(mat_a, mat_b):
    if mat_a.shape == mat_b.shape:
        mat_diff = mat_a - mat_b
    else:
        mat_diff = None
    return mat_diff
```

```python
#this function returns the multiplication of two matrices
def fn_mat_mul(mat_a, mat_b):
    if mat_a.shape[1] == mat_b.shape[0]:
        mat_mul = np.dot(mat_a, mat_b)
    else:
        mat_mul = None
    return mat_mul
```

```python
#this functions returns the subset of given matrix using row,column indexex
def fn_subset_mat(mat, r1, c1, r2, c2):
    
    if (r1>-1) and (c1>-1) and (r1<r2) and (c1<c2) and r2<mat.shape[0] and c2<mat.shape[1]:    
        res = mat[r1:r2, c1:c2]
    else:
        res = None
    return res
```

```python
#this is main function
if __name__ == "__main__":
    np.random.seed(3);
    ip_mat_a = np.random.randint(1, 20, size=(3, 3)); print("A=\n",ip_mat_a)
    ip_mat_b = np.random.randint(1, 20, size=(3, 3)); print("B=\n",ip_mat_b)
    ip_mat_c = np.random.randint(1, 20, size=(5, 5)); print("C=\n",ip_mat_c)
    res_sum = fn_mat_sum(ip_mat_a, ip_mat_b)
    res_diff = fn_mat_diff(ip_mat_a, ip_mat_b)
    res_mul = fn_mat_mul(ip_mat_a, ip_mat_b)
    res_subset_mat = fn_subset_mat(ip_mat_c, r1=1, c1=1, r2=3, c2=3)
    print("Sum:\n", res_sum)
    print("Diff:\n", res_diff)
    print("Mult:\n", res_mul)
    print("Subset:\n",res_subset_mat)
```
